Найдите ответы на вопросы или выполните действия с предложенным датасетом 

Данные о преступлениях, совершенных женщинами в Индии, с демографическими подробностями.

Файл: `CrimesOnWomenData.csv`

```
State - название штата/региона
Year - год записи
Rape - случаи изнасилования
K&A - случаи похищения 
DD - смерти из-за приданого
AoW - нападение на женщин
AoM - нападение на женскую скромность(modestity)
DV - бытовое насилие
WT - торговля женщинами
```

````python
import pandas as pd
````

````python
# Load dataset here
from io import StringIO
df = pd.read_csv('CrimesOnWomenData.csv')

print(df)
````

1. За какие года представлен набор данных?

````python
years = df['Year'].unique()

print("Годы, представленные в наборе данных: {", years[0], "-", years[len(years) - 1], "}")
````

2. Выявите район с самым высоким количеством убийств женщин

````python
result = df.groupby('State')['DD'].sum()

district_with_max_homicides = result.idxmax()
max_homicides = result.max()

print(f"Район с самым высоким количеством смертeй из-за приданого (?): {district_with_max_homicides} ({max_homicides} случаев)")
````

3. Какое среднее число изнасилований женщин в индии наблюдалось в указанные года?

````python
average_rapes = df[df['Year'].isin([2001, 2002, 2003])]['Rape'].mean()

print(f"Среднее число изнасилований женщин в Индии в указанные года: {average_rapes:.2f}")
````

4. Выявите район с минимальным числом похищений за период 2001-2003 года

````python
min_kidnappings_district = df[df['Year'].isin([2001, 2002, 2003])].groupby('State')['K&A'].sum().idxmin()
min_kidnappings_count = df[df['Year'].isin([2001, 2002, 2003])].groupby('State')['K&A'].sum().min()

print(f"Район с минимальным числом похищений (2001-2003): {min_kidnappings_district} ({min_kidnappings_count} случаев)")
````

5. Перечислите районы в которых наблюдались наиболее частые случае домашнего насилия за период с 2005 по 2012 года

````python
domestic_violence = df[df['Year'].isin(range(2005, 2013))].groupby('State')['DV'].sum()

most_frequent_domestic_violence = domestic_violence.sort_values(ascending=False)

print(most_frequent_domestic_violence[:10])
````

6. Выведите названия районов в порядке не возрастания случаев торговли женщинами за 2020 год

````python
trafficking_cases_2020 = df[df['Year'] == 2020].groupby('State')['WT'].sum()

sorted_trafficking_cases = trafficking_cases_2020.sort_values(ascending=False)

print(sorted_trafficking_cases)
````

Перейдем к следующему набору данных.

Файл: `CustomerPurchasingBehaviors.csv`

Этот набор данных содержит информацию о профилях клиентов и их покупательском поведении. Данные содержат столбцы, содержащие идентификатор пользователя, возраст, годовой доход, сумму покупки, показатель лояльности (в разбивке по классам), регион и частоту покупок. Он предназначен для анализа тенденций сегментации клиентов и лояльности и может быть использован для различных задач машинного обучения и анализа данных, связанных с поведением клиентов и исследованиями рынка.

```
user_id - ID клиента

annual_income - Годовой доход клиента (в долларах США).
purchase_amount - Общая сумма покупок, совершенных клиентом (в долларах США).
loyalty_score - Показатель лояльности клиента (значение от 0 до 100).

purchase_frequency - Частота покупок клиентов (количество раз в год).
```

````python
# Load dataset here
file_path = 'CustomerPurchasingBehaviors.csv' 
data = pd.read_csv(file_path)
print(data[:10])
````

8. Какой средний годовой доход у клиентов с уровнем лояльности больше 50 живущих на севере

````python
north_loyal_customers = data[(data['region'] == 'North') & (data['loyalty_score'] > 50)]

average_income_north = north_loyal_customers['annual_income'].mean()
print(f"Средний годовой доход у клиентов с уровнем лояльности больше 50, живущих на севере: {average_income_north}")

north_loyal_customers1 = data[(data['region'] == 'North') & (data['loyalty_score'] > 5.0)]
average_income_north1 = north_loyal_customers1['annual_income'].mean()
print(f"Средний годовой доход у клиентов с уровнем лояльности больше 5.0, живущих на севере: {average_income_north1}")
````

9. Выведите 25 и 75 перцентили уровня лоальности клиентов живущих на западе с частотой покупок более 20 раз в год

````python
west_frequent_customers = data[(data['region'] == 'West') & (data['purchase_frequency'] > 20)]

percentiles = west_frequent_customers['loyalty_score'].quantile([0.25, 0.75])
print(f"25-й : {percentiles[0.25]}, 75-й : {percentiles[0.75]}")
````

10. Отличается ли медианный возраст клиентов с уровнем лояльности более 70 и менее 10 для жителей Востока и Севера.

````python
east_loyal_customers = data[data['region'] == 'East']
median_age_east_high_loyalty = east_loyal_customers[east_loyal_customers['loyalty_score'] > 7.0]['age'].median()
median_age_east_low_loyalty = east_loyal_customers[east_loyal_customers['loyalty_score'] < 1.0]['age'].median()

# Фильтрация данных для Севера
north_loyal_customers = data[data['region'] == 'North']
median_age_north_high_loyalty = north_loyal_customers[north_loyal_customers['loyalty_score'] > 7.0]['age'].median()
median_age_north_low_loyalty = north_loyal_customers[north_loyal_customers['loyalty_score'] < 1.0]['age'].median()

print(f"Медианный возраст клиентов Востока с высоким уровнем лояльности (>7.0): {median_age_east_high_loyalty}")
print(f"Медианный возраст клиентов Востока с низким уровнем лояльности (<1.0): {median_age_east_low_loyalty}")
print(f"Медианный возраст клиентов Севера с высоким уровнем лояльности (>7.0): {median_age_north_high_loyalty}")
print(f"Медианный возраст клиентов Севера с низким уровнем лояльности (<1.0): {median_age_north_low_loyalty}")
````

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

1. Постройте диаграмму которая отражает средний уровень лояльности во всех регионах

````python
import matplotlib.pyplot as plt

average_loyalty_by_region = data.groupby('region')['loyalty_score'].mean()

average_loyalty_by_region.plot(kind='bar', color='skyblue')
plt.title('Средний уровень лояльности по регионам')
plt.xlabel('Регион')
plt.ylabel('Средний уровень лояльности')
plt.xticks(rotation=45)
plt.show()
````

2. Постройте гистограмму распределения зарплат клиентов на Севере.

````python
north_customers = data[data['region'] == 'North']

plt.hist(north_customers['annual_income'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Распределение зарплат клиентов на Севере')
plt.xlabel('Зарплата')
plt.ylabel('Частота')
plt.show()
````

3. Сравните уровни зарплат людей на севере и юге.

````python
north_south_customers = data[data['region'].isin(['North', 'South'])]

import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x='region', y='annual_income', data=north_south_customers)
plt.title('Сравнение уровней зарплат на Севере и Юге')
plt.xlabel('Регион')
plt.ylabel('Зарплата')
plt.show()
````

4. Постройте box plot отображающий распределение уровня лояльности для людей до 20 лет, от 20 до 40, от 40 до 60 и старше 60 лет

````python
bins = [0, 20, 40, 60, float('inf')]
labels = ['до 20', '20-40', '40-60', 'старше 60']
data['age_group'] = pd.cut(data['age'], bins=bins, labels=labels)

plt.figure(figsize=(10, 6))
sns.boxplot(x='age_group', y='loyalty_score', data=data)
plt.title('Распределение уровня лояльности по возрастным группам')
plt.xlabel('Возрастная группа')
plt.ylabel('Уровень лояльности')
plt.show()
````
